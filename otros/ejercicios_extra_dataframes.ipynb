5. Diccionario de vinos a DataFrame de vinos. Se dispone de un diccionario de vinos dV donde las 
claves son nombres de vinos y los valores para cada vino son una lista con la información de su 
denominación de origen, la añada y su precio.  

Crear un DataFrame a partir del diccionario dV que tenga en sus columnas 'Nombre', 
'Origen', 'Añada', 'Precio' y los datos incluyan las claves y valores de las listas de vinos. 
Usar colV e indV para las etiquetas de columns e index.

```python
dV = {'Muga':['Rioja', 2012, 8.5],'Petit caus': ['Penedes',2014, 6.75], 
     'Viña vial': ['Rioja',2016,9.5], 'Lavia': ['Jumilla',2011,12.5], 
     'Bach': ['Penedes',2015, 4.5], 'Pazo B': ['Rias Baixas', 2016,13.0]} 
colV = ['Nombre','Origen','Añada','Precio'] 
indV = ['V1','V2','V3','V4','V5','V6']

def adjoint_dictionary(dic):
        
    adjoint = {
    'Nombre': [],
    'Origen': [],
    'Añada': [],
    'Precio': [] }
    
    for name, values in dic.items():
        adjoint['Nombre'].append(name)
        adjoint['Origen'].append(values[0])
        adjoint['Añada'].append(values[1])
        adjoint['Precio'].append(values[2])
    
    return adjoint

df=pd.DataFrame(adjoint_dictionary(dV), columns=colV, index=indV)

df
```
